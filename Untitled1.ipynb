```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/content/salary_data.csv')
```

```python
X = data['YearsExperience'].values
y = data['Salary'].values
```

```python
m = 0 #gradient
b = 0 #bias
learning_rate = 0.01
epochs = 1000
```

```python
n = len(X)  # epoch
for _ in range(epochs):
    y_pred = m * X + b  # y = mx + b
    # gradient hisoblash
    D_m = (-2/n) * sum(X * (y - y_pred))  # m
    D_b = (-2/n) * sum(y - y_pred)        # b
    m = m - learning_rate * D_m
    b = b - learning_rate * D_b

print(" gradient (m):", m)
print(" bias (b):", b)
```

```python
plt.plot(X, y_pred, color='purple', marker='o', linestyle='-', label='Tahminiy oyliklar')
plt.xlabel('Tajriba (yillar)')
plt.ylabel('Prognoz qilingan oylik')
plt.title('Tahminiy oylik ish haqining grafigi')
plt.legend()
plt.show()
```
